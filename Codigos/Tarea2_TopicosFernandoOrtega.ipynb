Para empezar se tiene que definir inicialmente el análisis de correspondencias simple, para ello debemos definir una tabla de correspondencias, la cual es cualquier tabla bidireccional cuyas casillas contienen alguna medida de correspondencia entre las filas y las columnas. La medida de correspondencia puede ser cualquier indicación de la similitud, afinidad, confusión, asociación o interacción entre las variables de fila y columna. Un tipo muy común de tabla de correspondencias es una tabla cruzada, donde las casillas contienen recuentos de frecuencia.  
El análisis de correspondencias simple  permite examinar la relación entre dos variables nominales gráficamente en un espacio multidimensional. Calcula las puntuaciones de fila y columna y genera gráficos basados en las puntuaciones. Las categorías que son similares entre sí aparecen cerca una de la otra en los gráficos. De esta forma, es fácil ver qué categorías de una variable son similares entre sí o qué categorías de las dos variables están relacionadas. El procedimiento Análisis de correspondencias también le permite ajustar puntos suplementarios en el espacio definido por los puntos activos.  
La interpretación del análisis de correspondencias en términos de distancias depende del método de normalización utilizado. El procedimiento Análisis de correspondencias se puede utilizar para analizar las diferencias entre las categorías de una variable o las diferencias entre las variables. Con la normalización predeterminada, analiza las diferencias entre las variables de fila y columna.


**El análisis de correspondencia múltiple (ACM)** es una técnica estadística utilizada para explorar las relaciones entre variables categóricas en conjuntos de datos multivariados. A diferencia del análisis de correspondencia simple, que se enfoca en dos variables categóricas, el ACM se aplica a más de dos variables categóricas simultáneamente.

El objetivo principal del ACM es representar gráficamente las asociaciones entre las categorías de las diferentes variables, buscando patrones y tendencias en los datos. A través de una representación gráfica en un espacio de baja dimensionalidad, el ACM ayuda a visualizar la estructura subyacente de las relaciones entre las variables categóricas.

El proceso típico del análisis de correspondencia múltiple implica los siguientes pasos:

1. Tabulación de datos: Se crea una tabla de contingencia que muestra la frecuencia de ocurrencia conjunta de las diferentes categorías de las variables.

2. Cálculo de las correspondencias simples: Se realiza un análisis de correspondencia simple para cada par de variables, generando una representación gráfica para cada par.

3. Extracción de coordenadas: Se extraen las coordenadas factoriales de las correspondencias simples.

4. Análisis conjunto de correspondencias: Se combina la información de las correspondencias simples en un solo análisis, integrando las dimensiones obtenidas de cada par de variables.

5. Interpretación de resultados: Se interpreta la representación gráfica resultante, identificando patrones, asociaciones y relaciones entre las categorías de las variables.



El análisis de correspondencia múltiple (ACM) es apropiado en diversas circunstancias cuando se trabaja con datos categóricos. Algunos escenarios en los que esta técnica puede resultar beneficiosa incluyen:  

1. Análisis de encuestas: Cuando se tienen datos de encuestas con múltiples preguntas categóricas, el ACM puede ayudar a explorar las relaciones entre las respuestas y a identificar patrones subyacentes en las preferencias o comportamientos de los encuestados.

2. Investigación de mercado: En estudios de mercado, donde se recopilan datos sobre preferencias de productos, marcas o servicios, el ACM puede revelar asociaciones y tendencias entre diferentes categorías, facilitando la toma de decisiones estratégicas.

3. Ciencias sociales: En investigaciones sociales, el ACM puede utilizarse para analizar relaciones entre variables categóricas, como opiniones políticas, orientación religiosa u otras características sociodemográficas.

4. Biología y genética: En genética y biología, el ACM puede aplicarse para analizar la relación entre diferentes especies, genotipos u otras variables categóricas en conjuntos de datos biológicos.

5. Psicología: En estudios psicológicos, el ACM puede ayudar a entender las asociaciones entre diferentes dimensiones categóricas, como personalidad, comportamiento o preferencias.

6. Estudios de consumidores: En la investigación de comportamientos de consumidores, el ACM puede proporcionar información sobre las relaciones entre preferencias de productos, elecciones de marca y otros factores relacionados con el consumo.


* <font color="#CD2304">**Describe cómo se relaciona el ACM con el PCA y en qué se diferencian.** </font>  

El Análisis de Correspondencia Múltiple (ACM) y el Análisis de Componentes Principales (PCA) son métodos estadísticos que comparten la meta de reducir la dimensionalidad de los datos, aunque se distinguen por sus aplicaciones y la naturaleza de las variables con las que operan.  

**Relación entre ACM y PCA:**  

* Reducción de dimensionalidad: Tanto el ACM como el PCA buscan representar la información original en un espacio de menor dimensión, preservando la variabilidad presente en los datos.

* Descomposición en componentes: Ambos métodos implican descomponer la matriz de datos original en componentes principales (PCA) o coordenadas factoriales (ACM), capturando la variabilidad más significativa en los datos.

**Diferencias entre ACM y PCA:**

* Tipo de datos: Mientras el PCA se aplica principalmente a datos numéricos, el ACM está diseñado específicamente para datos categóricos. En el PCA, se exploran relaciones lineales entre variables cuantitativas, mientras que en el ACM se investigan relaciones entre categorías de variables cualitativas.

* Naturaleza de las variables: En el PCA, las variables son numéricas y continuas, utilizando la distancia euclidiana en el cálculo de componentes. En el ACM, las variables son categóricas, y la métrica de distancia se basa en frecuencias de co-ocurrencia de categorías.

* Interpretación de resultados: Aunque las componentes principales del PCA suelen tener interpretaciones directas, las coordenadas factoriales del ACM pueden ser más desafiantes de interpretar, ya que representan asociaciones entre categorías de variables.

* Aplicaciones específicas: Mientras el PCA se usa comúnmente en análisis de datos numéricos, como finanzas y biología, el ACM es especialmente valioso en análisis de datos categóricos, como encuestas y estudios de mercado.


* <font color="#CD2304">**Explica para qué se utiliza una tabla disyuntiva completa en el contexto de ACM.** </font>  

La Tabla Disyuntiva Completa es una tabla de n individuos por p categorías, en donde se utiliza un código indicador para cada categoría (1 si el individuo la asume, 0 si no), denominado disyuntivo completo, por sus propiedades.  

En el contexto del Análisis de Correspondencia Múltiple (ACM), una tabla disyuntiva completa es una herramienta crucial que se utiliza para representar las relaciones entre múltiples variables categóricas. La tabla disyuntiva completa es una tabla de contingencia expandida que muestra la co-ocurrencia de todas las combinaciones posibles de categorías de las variables consideradas en el análisis.  

**carácteristicas:**  

* Representación exhaustiva: La tabla disyuntiva completa garantiza que se incluyan todas las combinaciones posibles de categorías de las variables en el análisis. Esto es esencial para capturar de manera exhaustiva las asociaciones y patrones entre las distintas categorías.

* Entrada principal para el análisis: La tabla disyuntiva completa sirve como la entrada principal para realizar el ACM. A partir de esta tabla, se calculan las coordenadas factoriales que representan las relaciones entre las categorías de las variables.

* Frecuencias de co-ocurrencia: La tabla disyuntiva completa contiene las frecuencias de co-ocurrencia de cada combinación de categorías, lo que permite al ACM evaluar la fuerza y la dirección de las asociaciones entre las variables.

* Reducción de dimensionalidad: A través del ACM, se busca reducir la alta dimensionalidad de la tabla disyuntiva completa a un espacio de menor dimensión, preservando la estructura de las relaciones entre las variables categóricas.

* Visualización y interpretación: El análisis de correspondencia múltiple utiliza las coordenadas factoriales derivadas de la tabla disyuntiva completa para crear representaciones gráficas que facilitan la interpretación de las relaciones entre las categorías. Estas representaciones visuales ayudan a identificar patrones y tendencias de manera más clara.

La construcción de una tabla disyuntiva completa es un paso importante en el proceso de preparación de datos para un Análisis de Correspondencia Múltiple (ACM).Para construirla se tienen que seguir los siguientes pasos:

* <font color="#CD2304">**Enumera y describe las limitaciones y desafíos al realizar un ACM.** </font>  

* Dificultad con Datos Continuos:  

El ACM está diseñado principalmente para datos categóricos, y su eficacia disminuye cuando se aplican a conjuntos de datos con variables continuas. Para datos mixtos (categóricos y continuos), se deben considerar métodos alternativos.  

* Sensibilidad a la Representación de Datos:  

La interpretación del análisis puede depender de la representación específica de la tabla disyuntiva completa, y pequeños cambios en la representación pueden afectar los resultados.  

* No Considera Relaciones Lineales Complejas:  

El ACM se centra en relaciones no lineales entre variables categóricas. Si las relaciones son predominantemente lineales, otras técnicas como el Análisis de Componentes Principales (PCA) pueden ser más apropiadas.  

* Sensibilidad a las Frecuencias Bajas:  

Las frecuencias bajas en la tabla disyuntiva completa pueden afectar la estabilidad y la confiabilidad del análisis. Se deben tener precauciones al interpretar asociaciones basadas en frecuencias bajas.  

* Interpretación Subjetiva:  

La interpretación de las coordenadas factoriales puede ser subjetiva y depende de la experiencia y el juicio del analista. La interpretación no siempre es directa, especialmente cuando hay muchas variables.  

* Sobreajuste :

En conjuntos de datos pequeños, existe el riesgo de sobreajuste, donde el modelo se ajusta demasiado a los datos de muestra y puede tener dificultades para generalizar a nuevos datos.  

* Altas Dimensiones:  

Cuando se trabaja con conjuntos de datos con un gran número de variables categóricas, la interpretación y visualización de los resultados pueden volverse desafiantes debido a la alta dimensionalidad.
 
* Dependencia de la Elección del Número de Dimensiones:  

La elección del número de dimensiones a retener puede ser subjetiva y afectar la interpretación de los resultados. 

* Limitaciones en Datos Longitudinales:  

El ACM es más adecuado para datos de tipo transversal y puede no ser la mejor opción para datos longitudinales, donde se estudia el cambio a lo largo del tiempo.



* <font color="#CD2304">**Identifica y explica las fórmulas matemáticas clave que se requieren en el ACM.**</font>  

1. Tablas de contingencia:  
Sea n el número total de individuos, J y K son las categorías de las dos variables categóricas.  
$X_{ijk}$ representa la frecuencia de individuos que tienen la categoría j en la variable 1 y la categoría k en la variable 2.  
2. Recuencias relativas  
$\Large f_{ijk}=\frac{x_{ijk}}{n}$  

3. Frecuencias Marginales  
Frecuencia marginal de la fila i:  $r_{i}=\Large\sum_{j=1}^{J}\Large\sum_{k=1}^{K}f_{ijk} $  

Frecuencia marginal de la columna k: $c_{k}=\Large\sum_{i=n}^{J}\Large\sum_{j=1}^{J}f_{ijk} $  

4. Matriz de chi cuadrado  

$\chi^{2}=\Large\sum_{i=1}^{n}\Large\sum_{j=1}^{j}\Large\sum_{k=1}^{k}\frac{(f_{ijk}-e_{ijk})^2}{e_{ijk}} $    
Donde 
$e_{ijk}$ es la frecuencia esperada para la celda $ijk$  

5. Descompisicion de valores singulares SVD de \chi^{2}  

La SVD de la matriz de chi-cuadrado proporciona los valores singulares que están relacionados con la varianza explicada en el espacio de dimensiones reducidas.  
$\Large\chi^{2} = U\Sigma V´$  

Done U y V son matrices de coordenadas factoriales y $\Sigma$ es una matriz diagonal con los valores singulares  

6. Puntuación de filas y columnas  
Puntuaciones de las filas en el espacio reducido: $\Large F=U\sqrt{\Sigma}$

Puntuaciones de las columnas en el espacio reducido: $\Large G=V\sqrt{\Sigma}$   
  
7. Inercia Total   
$\Large I_{total}= \sum_{i=i}^{r}\sigma_{i}^2$  

Donde $\sigma_{i}^2$ son los valores singulares obtenidos de la SVD.  
8. Inercia Explicada  
$\Large I_{explicada}= \sum_{i=i}^{k}\sigma_{i}^2$   

Donde k es el número de dimensiones retenidas  
9. Inercia Total   
$\Large I_{relativa}= \frac{I_{explicada}}{I_{total}}$  



10. Coordenadas Factoriales:  
Las coordenadas factoriales son los valores numéricos que representan la posición de las categorías en el espacio de factores. Para calcular las coordenadas factoriales $f_{ij}$ de la categoria $i$ de la variable $X$ en la dimension $k$.Se utiliza la siguiente fórmula:

$f_{ij}=\sqrt{\lambda_{k}} . u_{ij} . \sqrt{P_{ij}}$  

Donde:  
$\lambda_{k}$ : es el valor propio asociado con la dimensión K  
$u_{ij}$ :es el elemento de la matriz U de la descomposición en valores singulares (SVD)
$P_{ij}$ : es la frecuencia relativa de la categoría i de la variable X

La interpretación de los gráficos resultantes de un Análisis de Correspondencia Múltiple (ACM) es crucial para comprender las relaciones entre las categorías de las variables analizadas. Los gráficos más comunes en ACM son los gráficos de coordenadas factoriales para las filas y las columnas.


* Gráfico de Coordenadas Factoriales:  
El gráfico de coordenadas factoriales visualiza la posición de las categorías de las variables (filas y columnas) en el espacio de dimensiones reducidas obtenido a través de la SVD en el ACM. Este tipo de gráfico es útil para identificar patrones, relaciones y asociaciones entre las categorías.  

**Interpretación**  

* Distancias entre Puntos:  

La distancia entre dos puntos en el gráfico refleja la similitud o asociación entre las categorías correspondientes. Puntos cercanos indican asociaciones fuertes, mientras que puntos alejados indican asociaciones débiles.


* Orientación de los Puntos:

La orientación de los puntos resalta la dirección de las asociaciones entre las categorías. Categorías con orientaciones similares están relacionadas positivamente, mientras que orientaciones opuestas indican relaciones negativas.


* Concentración de Puntos:

La concentración de puntos alrededor de una región específica del gráfico sugiere la existencia de patrones o grupos de asociación entre las categorías.

* <font color="#CD2304">**Discute el papel que juega la varianza en los métodos de ACM y PCA.**</font>  

La varianza juega un papel crucial en los métodos de Análisis de Correspondencia Múltiple (ACM) y Análisis de Componentes Principales (PCA), ya que estos métodos están diseñados para maximizar la varianza en ciertas direcciones del espacio de datos.

* Análisis de Correspondencia Múltiple (ACM):  

Inercia Total:  

En el ACM, se busca explicar la inercia total presente en la tabla de contingencia de las variables categóricas. La inercia total representa la variabilidad total en los datos.  

Valores Singulares:  

Durante la Descomposición en Valores Singulares (SVD), se obtienen los valores singulares, que son las raíces cuadradas de los autovalores de la matriz de chi-cuadrado. Los valores singulares representan la cantidad de varianza explicada por cada dimensión del espacio de dimensiones reducidas.  

Coordinadas Factoriales:  

Las coordenadas factoriales (obtenidas de las matrices U y V en la SVD) indican cómo se distribuye la varianza entre las categorías de las variables en el espacio de dimensiones reducidas.  

* Análisis de Componentes Principales (PCA):  

Varianza Total:  
 
En PCA, se busca encontrar las direcciones (componentes principales) a lo largo de las cuales la varianza de los datos es máxima.

Matriz de Covarianza:

PCA se basa en la matriz de covarianza de las variables. La covarianza es una medida de la variabilidad conjunta entre dos variables. La matriz de covarianza resume la variabilidad y las relaciones lineales entre las variables originales.

Autovectores y Autovalores:  

Al resolver el problema de los autovectores y autovalores de la matriz de covarianza, se obtienen los autovectores (que definen las direcciones de las componentes principales) y los autovalores (que indican la cantidad de varianza explicada por cada componente principal).  

Proyección de Datos:  

Las componentes principales se utilizan para proyectar los datos originales en un espacio de dimensionalidad reducida, maximizando así la varianza total retenida en las nuevas dimensiones.

**Describe la representación matricial de la varianza en ACM y PCA.**

En Análisis de Correspondencia Múltiple (ACM) y Análisis de Componentes Principales (PCA), la varianza se puede expresar matricialmente a través de la matriz de covarianza o la matriz de inercia, según el contexto.  

* Representación Matricial de la Varianza en ACM:  
Matriz de Chi-Cuadrado :  

1. La matriz de chi-cuadrado es central en ACM y representa la relación entre las categorías de las variables. La inercia total en ACM se relaciona con la varianza total de esta matriz.

$\chi^{2}=\Large\sum_{i=1}^{n}\Large\sum_{j=1}^{j}\Large\sum_{k=1}^{k}\frac{(f_{ijk}-e_{ijk})^2}{e_{ijk}} $  

Donde $f_{ijk}$ es la refuencia observaoda, $e_{ijk}$ es la frecuenia esperada y la suma se realiza sobre todas las categorías y observaciones.  

2. Descomposición en Valores Singulares (SVD):  
La SVD de la matriz de chi-cuadrado proporciona los valores singulares que están relacionados con la varianza explicada en el espacio de dimensiones reducidas.  
$\Large\chi^{2} = U\Sigma V´$  

Done U y V son matrices de coordenadas factoriales y $\Sigma$ es una matriz diagonal con los valores singulares

* Representación Matricial de la Varianza en PCA:  

1. Matriz de covarianza:  
En PCA, la matriz de covarianza es clave para describir la varianza y las relaciones entre las variables originales.  
$Cov = \Large\frac{1}{n}=\Large\sum_{i=1}^{n}(X_{i}-\bar{X})(X_{i}-\bar{X})´$  
Donde $X_{i}$ es el vector de variables para la i-ésima observación, $\bar{X}$ es el vector de medias de las variables y la suma se realiza sobre todas las observaciones.  
2. Descomposición Espectral:
La descomposición espectral de la matriz de covarianza  proporciona los autovectores  y los autovalores, que están relacionados con las direcciones y la cantidad de varianza explicada en PCA.  

$C = \Large V Λ Λ´$  
Donde V es la matriz de autovectores y Λ es la matriz diagonoal de autovalores.

## Realiza el análisis de correspondencia múltiple (MCA)

```python
# importamos la libreria prince la cual nos ayudará al análisis de correspondencias
import prince
```

```python
# Ajustamos nuestro moedelo de mca con 6 componentes dado que tenemos 6 categorías

mca = prince.MCA(n_components=6, n_iter=25, engine='sklearn', random_state=1995)
mca = mca.fit(set_categorico)
```

Analizamos primeramente la distribución de renglones

```python
cordenadas_mca = mca.row_coordinates(set_categorico)   # guardamos las coordenadas  factoriales de los renglones
```

#### Veamos un resumen del modelo de ACM generado

```python
eigenvalores = mca.eigenvalues_


porcent_varianza = (eigenvalores / np.sum(eigenvalores)) * 100


varianza_acumulada = np.cumsum(porcent_varianza)


summary = pd.DataFrame({
    'Eigenvalue': eigenvalores,
    '% Variance': porcent_varianza,
    '% Cumulative Variance': varianza_acumulada
})

summary
```

```python
# Crea un gráfico de dispersión usando las coordenadas de los individuos.

plt.figure(figsize=(12, 10))
sns.scatterplot(x=-1*cordenadas_mca[0], y=cordenadas_mca[1], alpha=0.8,color="orange", label='Razas')
plt.grid()

# Añade líneas de referencia para los ejes
plt.axhline(0, color='red', linewidth=1, linestyle='--')
plt.axvline(0, color='red', linewidth=1, linestyle='--')

plt.title("Representación individual de razas de perros",fontsize=16)
plt.xlabel(f"Dim 1 ({round(porcent_varianza[0],1)} %)")
plt.ylabel(f"Dim 2 ({round(porcent_varianza[1],1)} %)")

# Ajusta el texto para evitar superposiciones
texts = [plt.text(-1*cordenadas_mca.iloc[i, 0]+.01, cordenadas_mca.iloc[i, 1]+.085, raza.capitalize()) for i, raza in enumerate(base["raza"])]
adjust_text(texts)

```

Este gráfico nos da un primer approach de la similitud entre los rengones, en este caso estamos capturando la información de la tabla de contingencia respecto a las 6 variables cualitativas, en dos dimensiones, las cuales equivalen a un 55 % de la varianza explicada. A partir de esto en los grupos podemos empezar a ver similitudes como los Peki y chihuahua, probablemente por tamaño, lo mismo podemos ver en razas grandes como el boxer, labrador y el dalmata.

```python
# Crea un gráfico de barras con Seaborn
plt.figure(figsize=(12, 10))


# Añade una línea roja punteada en un umbral específico
plt.axhline(y=7.14, linestyle='--', color='red', label='Threshold')

sns.lineplot(x=list(range(0, len(eigenvalores) )), y=porcent_varianza, marker="o")

sns.barplot(x=list(range(1, len(eigenvalores) + 1)), y=porcent_varianza ,alpha=0.7,palette="inferno")

# Añade etiquetas para cada barra
for i, value in enumerate(porcent_varianza):
    plt.text(i, value + 0.2, f'{value:.2f}%', ha='center', va='bottom')

# Ajusta el título y etiquetas
plt.title('Screeplot - Percentage of Variance')
plt.xlabel('Dimensions')
plt.ylabel('% Variance')
plt.xticks(range(0, len(eigenvalores)))

# Muestra el gráfico con límite ajustado en el eje y
plt.ylim(0, max(15, np.max(porcent_varianza) + 1))  # Ajusta el límite del eje y
plt.legend()
plt.show()
```

Ya con los porcentajes de varianza se puede definir la cantidad de componentes pricipales, en este caso los 4 primeros superan el 91%.  

Al hacer el análisis completo vemos que en general se usan 2 componentes para representar en dos dimensiones, y en nuestro caso particular vemos como estas dos dimensiones nos aportan la mayor cantidad de varianza explicada, con esto vemos que para fines teoricos con los 6 componentes podemos generar mas del 90 % de la varianza de los datos.

El Análisis de Correspondencias Múltiples (ACM) se revela como una herramienta valiosa en el análisis exploratorio de datos multivariantes, permitiendo la visualización y comprensión de las relaciones complejas entre múltiples variables categóricas. En este estudio, hemos observado cómo el ACM ha contribuido significativamente a la identificación de patrones y asociaciones entre las categorías de nuestras variables, facilitando la interpretación de la estructura subyacente de los datos.

Las dimensiones principales extraídas a través del ACM han proporcionado una representación visual efectiva de la información contenida en nuestras variables, destacando las tendencias y relaciones más relevantes. Estas dimensiones no solo simplifican la complejidad de los datos, sino que también ofrecen una base sólida para la formulación de hipótesis y la generación de nuevas preguntas de investigación.

Este enfoque ha demostrado ser particularmente útil en la exploración de datos categóricos, donde otras técnicas estadísticas podrían ser menos apropiadas.

No obstante, es crucial reconocer las limitaciones inherentes al ACM, como su sensibilidad a la elección de las categorías activas e inactivas, así como la necesidad de interpretar las dimensiones en función de la naturaleza específica de los datos y el contexto del estudio.

En resumen, el ACM se posiciona como una herramienta valiosa en la investigación multivariante, proporcionando una perspectiva única y facilitando la interpretación de relaciones complejas entre variables categóricas. Su aplicación efectiva puede no solo mejorar nuestra comprensión de la estructura de los datos, sino también orientar investigaciones futuras y aplicaciones prácticas en diversas disciplinas.
