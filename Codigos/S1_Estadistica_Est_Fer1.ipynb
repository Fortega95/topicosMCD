Calcular el promedio aritmético de los valores en un DataFrame $$\bar{x} = \frac{1}{n} \sum_{i=1}^{n} x_i$$

Calcular la mediana de los valores en un DataFrame $$\text{Mediana} = 
\begin{cases} 
x_{(n+1)/2} & \text{si } n \text{ es impar} \\
\frac{x_{(n/2)} + x_{(n/2) + 1}}{2} & \text{si } n \text{ es par} 
\end{cases}$$

```python
data.total_rooms.median()
```

Calcular la moda o modas de los valores en un DataFrame o Serie, siendo la moda el valor o valores que aparecen con mayor frecuencia. $$\text{Moda} = \{x_i : \text{frecuencia de } x_i \text{ es máxima}\}$$
 

```python
data.total_rooms.mode()
```

En este contexto el análisis de correlacion nos ayudará para encontrar relaciones lineales entre variables con lo que podremos reducir de dimensionalidad, eliminar datos redundantes asi como bajar la complejidad de la base descartando variables que no sean importantes en el modelo que queramos implementar, ya sea para predecir o clasificar.
Igualmente cuando solo tratamos de hacer un análisis descriptivo, las correlaciones nos sirven para dar contexto y resolver preguntas que salgan respecto al comportamiento de los datos, lo cual nos permite darle una mejor comprension a nuestra base asi como a la problematica y los factores que la afectan directamente.
